```python
name = input("What is your name?: ")
numberqscore = int(input("How many quiz scores do you wanna input?: "))

def avg(total, count):
    return total / count

total = 0

if numberqscore <= 0 or numberqscore > 100:
    print("Invalid input")
else:
    for i in range(numberqscore):
        score = float(input("Enter score: "))
        total = total + score

    average = avg(total, numberqscore)

    print("RESULT")
    print("Name:", name)
    print("Average:", average)

    if average >= 75:
        print("Result: PASSED")
    else:
        print("Result: FAILED")
```
